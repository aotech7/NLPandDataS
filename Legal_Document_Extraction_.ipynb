```python
# pip install spacy library for text processing
!pip install spacy
# installs small version of an English language model

#!python -m download en_core_web_sm


#import other libraries
import spacy
import pandas as pd
from google.colab import files

#import files from desktop

def upload_read_nda():
    uploaded_f = files.upload()
    if not uploaded_f:
      print("No file was uploaded.")
      return None

    file_name = next(iter(uploaded_f))

  # utf encodings

    encodings = ['utf-8', 'ascii', 'latin-1', 'utf-16']


    for encoding in encodings:
        try:
          content = uploaded_f[file_name].decode(encoding)
          print(f"Successfully read file using {encoding} encoding.")
          return content

        except UnicodeDecodeError:
            continue


    print("Unable to read file. Is it a text file? ")
    return None

print("Please upload your NDA file:")
nda_text = upload_read_nda()

if nda_text is not None:
      print("File uploaded adn read successfully.")
      print("First 100 characters of file: ")
      print(nda_text[:100])
else:
        print("Failed to read the file")


```
